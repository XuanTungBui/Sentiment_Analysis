```python
import glob

n = glob.glob("./data_train_do_an/data_train/train/neg/*.txt")
p = glob.glob("./data_train_do_an/data_train/train/pos/*.txt")

def map_first_lines(file_list):
    for file in file_list:
        with open(file, 'r',encoding="utf8") as fd:
            yield fd.readline()


neg = [f for f in map_first_lines(n)]
pos = [f for f in map_first_lines(p)]
```

```python
print(len(neg))
print(len(pos))
```

```python
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.DataFrame(neg, columns = ['comment'])
label = np.ones(20000,dtype='int')
df['label'] = label
df
```

```python
df1 = pd.DataFrame(pos, columns = ['comment'])
label1 = np.zeros(20000,dtype='int')
df1['label'] = label1
df1
```

```python
data = pd.concat([df,df1],ignore_index = True)
data
```

```python
df2 = pd.read_csv('train.csv')
df2
```

```python
df2 = df2.iloc[:,1:]
df2.head()
```

```python
data1 = pd.concat([data,df2],ignore_index = True)
data1
```

```python
data1['comment'] = data1['comment'].astype('str')
mask = (data1['comment'].str.len() < 300)
data1 = data1.loc[mask]
data1
```

```python
data1 = data1.sample(frac = 1)
data1
```
